## Обучение модели

<div style="border-radius: 15px; border: 1px solid grey; padding: 15px;">
Начнем с того, что код в этом разделе запускался в облаке с GPU, поэтому оформлен не как ячейка с кодом, а как код в текстовой ячейке.
    
Теперь переходим к модели - код выполняемый в облаке должен был быть офрмлен в виде несколльких функций:
    
    1-я и 2-я функции load_train и load_test формируют тензоры обучающей и тестовой выборок. Разделение по выбркам производилось в соотношение 3/1. Прозведен рескайл цветов что бы значения находились в диапазоне от 0 до 1. Изображения остались в исходном размере 256х256, а размер батча соствил 16 изображений. Так же для тестовой выбрки был добавлен параметыр горизонтально поворота изображений для увеличения количества тренировочных данных.
    
    Функция create_model создает модель. В базе нашей модели лежить нейросеть ResNet с 50-ю слоями у которой убрана верхушка. В качестве финальных слоев (верхушки) были последовтельно добавлены слои пуллига всей картинки, приведения каждой картинки в одномерный массив и два полносвязных слоя со 101 нейроном и 1-м нейроном в финале, функцие активации ReLU. При компеляции основная метрика и лосс функция были выбраны различными а мименно MAE и MSE соответственно, плюсь оптимизационной функцие выбрана Adam с достаточно низким шангом обучения в 0.0001.
    
    И последенй идет функция train_model запускающая процес собучения сети в котором задано 30 эпох обучения и параметр вывода результатов обучения на экран.
    
Далее перейдем уже к анализу вывода обученяи модели.
</div>



```python
from tensorflow.keras.layers import Flatten, Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.resnet import ResNet50
import pandas as pd


def load_train(path):
    labels = pd.read_csv(path + 'labels.csv')
    
    train_datagen = ImageDataGenerator(
        validation_split=0.25,
        horizontal_flip=True,
        rescale=1. / 255)
    
    train_gen_flow = train_datagen.flow_from_dataframe(
        dataframe=labels,
        directory=path + 'final_files/',
        x_col='file_name',
        y_col='real_age',
        target_size=(256, 256),
        batch_size=16,
        class_mode='raw',
        subset='training',
        seed=12345)
    
    return train_gen_flow


def load_test(path):
    labels = pd.read_csv(path + 'labels.csv')
    
    test_datagen = ImageDataGenerator(
        validation_split=0.25,
        rescale=1. / 255)
    
    test_gen_flow = test_datagen.flow_from_dataframe(
        dataframe=labels,
        directory=path + 'final_files/',
        x_col='file_name',
        y_col='real_age',
        target_size=(256, 256),
        batch_size=16,
        class_mode='raw',
        subset='validation',
        seed=12345)

    return test_gen_flow


def create_model(input_shape):
    optimizer = Adam(lr=0.0001)
    backbone = ResNet50(input_shape=input_shape,
                        weights='imagenet',
                        include_top=False)
    model = Sequential()
    model.add(backbone)
    model.add(GlobalAveragePooling2D())
    model.add(Flatten())
    model.add(Dense(101, activation='relu'))
    model.add(Dense(1, activation='relu'))

    model.compile(loss='mse', optimizer=optimizer, metrics=['mae'])
    return model


def train_model(model, train_data, test_data, batch_size=None, epochs=30,
               steps_per_epoch=None, validation_steps=None):
    model.fit(train_data,
              validation_data=test_data,
              batch_size=batch_size, epochs=epochs,
              steps_per_epoch=steps_per_epoch,
              validation_steps=validation_steps,
              verbose=2, shuffle=True)
    return model
```
