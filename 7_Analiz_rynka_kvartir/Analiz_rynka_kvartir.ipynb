## Предобработка данных

### Обработка пропусков

````python
df['floors_total'] = df['floors_total'].fillna(df['floor'] + 1)

df['living_area'] = df['living_area'].fillna(df['total_area'] - df['kitchen_area'])

df['balcony_total'] = df['balcony_total'].fillna(0)

df['parks_around3k_total'] = df['parks_around3k_total'].fillna(0)

df['ponds_around3k_total'] = df['ponds_around3k_total'].fillna(0)

df.loc[df['airports_nearest_meters'] == 0, ['airports_nearest_meters']] = np.nan

df['exposition_days'] = df['exposition_days'].fillna(-1)

df['is_apartment_type'] = df['is_apartment_type'].fillna(False)

df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
````

**>>>**

   `floors_total` - пропуски этажности здания заполнены текущим значением этажа квартиры + 1, значений не много, а вероятность, что текущий этаж является последним значительно меньше промежуточного этажа. Появление этих пропусков вероятно обусловлено человеческим фактором при заполнении анкеты, неплохо бы сделать такое поле обязательным к заполнению.
    
   `living_area` и `kitchen_area` - 439 значений жилой пелощади высчитано вычетанием площади кухни из общей площади, это не всегда так, но такая жилая площадь гораздо ближе к истине, чем пропуск. Обратная оперяция вычисления площади кухни из жилой и общей считаю невозможной, так как кухни может не быть вовсе, а не жилая площадь будет, например если продается комната с балконом. Появление этих пропусков вероятно обусловлено человеческим фактором при заполнении анкеты отсутсвием зоны кухни.
    
   `balcony_total` - пропуски замененры на "0", скорее всего человек не заполнил графу в силу отсутствия балкона.

Появление этих пропусков вероятно обусловлено человеческим фактором при заполнении анкеты.

Появление этих пропусков вероятно появились по технической причине, данные небыли подгружены или висчитаны в момент прекращения публикации, ещё возможен вариант, что публикация не была снята на момент выгрузки, но такие данные не должны попадать в выгрузку.

### Замена типов данных

````python
try:
    df['last_price'] = df['last_price'].astype('int')
except:
    print('Столбец last_price содержит данные кторые невозможно привести к целочисленным.')
    
try:
    df['floors_total'] = df['floors_total'].astype('int')
except:
    print('Столбец floors_total содержит данные кторые невозможно привести к целочисленным.')
    
try:
    df['balcony_total'] = df['balcony_total'].astype('int')
except:
    print('Столбец balcony_total содержит данные кторые невозможно привести к целочисленным.')

try:
    df['airports_nearest_meters'] = df['airports_nearest_meters'].astype('Int64')
except:
    print('Столбец airports_nearest_meters содержит данные кторые невозможно привести к Int64.')
    
try:
    df['city_center_nearest_meters'] = df['city_center_nearest_meters'].astype('Int64')
except:
    print('Столбец city_center_nearest_meters содержит данные кторые невозможно привести к Int64.')

try:
    df['parks_around3k_total'] = df['parks_around3k_total'].astype('int')
except:
    print('Столбец parks_around3k_total содержит данные кторые невозможно привести к целочисленным.')

try:
    df['parks_nearest_meters'] = df['parks_nearest_meters'].astype('Int64')
except:
    print('Столбец parks_nearest_meters содержит данные кторые невозможно привести к Int64.')

try:
    df['ponds_around3k_total'] = df['ponds_around3k_total'].astype('int')
except:
    print('Столбец ponds_around3k_total содержит данные кторые невозможно привести к целочисленным.')

try:
    df['ponds_nearest_meters'] = df['ponds_nearest_meters'].astype('Int64')
except:
    print('Столбец ponds_nearest_meters содержит данные кторые невозможно привести к Int64.')

try:
    df['exposition_days'] = df['exposition_days'].astype('int64')
    
except:
    print('Столбец exposition_days содержит данные кторые невозможно привести к Int64.')
````

В четырех члучаях был выбран формат 'Int64', так как он позволяет обрабатывать пропуски.
    Пропуски в `exposition_days` обрабатываться отказывались в силу не врозможности безопасной конвертации, пришлось временно заменить пропуски на -1 и вернуть NaN после.

### Тип этажа квартиры

````python
def flo (row):
    floor = row['floor']
    top_floor = row['floors_total']
    if floor == 1:
        return 'Первый'
    elif floor == top_floor:
        return 'Последний'
    else:
        return 'Другой'
        
df['floor_type']=df.apply(flo, axis=1)
    
df['floor_type'].value_counts()
````
